```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Diabetes Classification.csv")

# Preview data
df.head()
```

```python
# Shape of dataset
print("Dataset Shape:", df.shape)

# Column info
df.info()

# Summary statistics
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure()
plt.hist(df['Age'], bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure()
df['Gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure()
df.boxplot(column='BMI', by='Diagnosis')
plt.title('BMI by Diabetes Diagnosis')
plt.suptitle('')
plt.xlabel('Diagnosis')
plt.ylabel('BMI')
plt.show()
```

```python
plt.figure()
plt.scatter(df['BMI'], df['Chol'])
plt.title('BMI vs Cholesterol')
plt.xlabel('BMI')
plt.ylabel('Cholesterol')
plt.show()
```
