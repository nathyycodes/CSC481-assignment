```python
# -----------------------------------
# 1. Import Libraries
# -----------------------------------
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------------
# 2. Load Dataset
# -----------------------------------
file_path = "healthcare_noshows_appointments.csv"
df = pd.read_csv(file_path)

# -----------------------------------
# 3. Initial Data Exploration
# -----------------------------------
print("Dataset Shape:")
print(df.shape)

print("\nDataset Info:")
df.info()

print("\nFirst 5 Rows:")
df.head()
```

```python
# -----------------------------------
# 4. Basic Statistics
# -----------------------------------
df.describe()
```

```python
# -----------------------------------
# 5. Visualization 1: Show-up vs No-show
# -----------------------------------
plt.figure()
df['Showed_up'].value_counts().plot(kind='bar')
plt.title("Appointment Attendance")
plt.xlabel("Showed Up")
plt.ylabel("Number of Appointments")
plt.show()
```

```python
# -----------------------------------
# 6. Visualization 2: Age Distribution
# -----------------------------------
plt.figure()
df['Age'].plot(kind='hist', bins=30)
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# -----------------------------------
# 7. Visualization 3: SMS Reminder vs Show-up Rate
# -----------------------------------
sms_show_rate = df.groupby('SMS_received')['Showed_up'].mean()

plt.figure()
sms_show_rate.plot(kind='bar')
plt.title("Show-up Rate by SMS Reminder")
plt.xlabel("SMS Received")
plt.ylabel("Show-up Rate")
plt.show()
```

```python
# -----------------------------------
# 8. Additional Analysis (Optional but Recommended)
# No-show rate by gender
# -----------------------------------
gender_show = df.groupby('Gender')['Showed_up'].mean()

plt.figure()
gender_show.plot(kind='bar')
plt.title("Show-up Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Show-up Rate")
plt.show()
```
